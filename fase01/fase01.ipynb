Fase 1 FIAP - Ignition Zero: O Início da Aurora

```python
import random

print("=== Sistema Ignition Zero - Verificação de Pré-Decolagem ===\n")

# Função para simular telemetria
def interpretar_telemetria():
    telemetria = {
        'temperatura_interna': random.uniform(-10, 60), # °C
        'temperatura_externa': random.uniform(-80, 40), # °C
        'integridade_estrutural': random.uniform(0, 1), # 0 a 1
        'nivel_energia': random.uniform(50, 100), # %
        'pressao_tanques': random.uniform(30, 100), # escala simulada
        'modulos_criticos': random.choice([0, 1]) # 1 = OK | 0 = FALHA
    }
    return telemetria


# Verificação dos parâmetros de segurança
def verificar_parametros(telemetria):

    if telemetria['temperatura_interna'] < -20 or telemetria['temperatura_interna'] > 50:
        return "DECOLAGEM ABORTADA - Temperatura interna fora do limite"

    if telemetria['integridade_estrutural'] < 0.8:
        return "DECOLAGEM ABORTADA - Integridade estrutural crítica"

    if telemetria['nivel_energia'] < 70:
        return "DECOLAGEM ABORTADA - Energia insuficiente"

    if telemetria['pressao_tanques'] < 40 or telemetria['pressao_tanques'] > 90:
        return "DECOLAGEM ABORTADA - Pressão irregular"

    if telemetria['modulos_criticos'] == 0:
        return "DECOLAGEM ABORTADA - Falha em módulos críticos"

    return "PRONTO PARA DECOLAR"


# Análise energética
def analisar_energetica(telemetria):

    capacidade_total = 100  # kWh (valor simulado)

    carga_atual = telemetria['nivel_energia'] / 100
    energia_disponivel = capacidade_total * carga_atual

    consumo_decolagem = 30  # kWh estimado
    perdas = energia_disponivel * 0.05  # 5% de perdas

    energia_final = energia_disponivel - consumo_decolagem - perdas

    resultado = f"""
--- Análise Energética ---
Energia disponível: {energia_disponivel:.2f} kWh
Consumo estimado: {consumo_decolagem} kWh
Perdas energéticas: {perdas:.2f} kWh
Energia restante após decolagem: {energia_final:.2f} kWh
"""
    return resultado


# Função principal
def main():

    telemetria = interpretar_telemetria()

    print("Dados da Telemetria:\n")
    for chave, valor in telemetria.items():
        print(f"{chave}: {valor:.2f}" if isinstance(valor, float) else f"{chave}: {valor}")

    print("\n-----------------------------")

    resultado = verificar_parametros(telemetria)
    print("\nResultado da verificação:")
    print(resultado)

    print(analisar_energetica(telemetria))

    print("Reflexão:")
    print("A automação em sistemas críticos exige responsabilidade ética e eficiência energética,")
    print("pois decisões computacionais impactam diretamente a segurança e o uso sustentável de recursos.")


# Execução
if __name__ == "__main__":
    main()
```
